## mean 'price_usd' per 'brand'

```python
autos["brand"].sort_values().unique()
```

```python
autos["price_usd"].groupby(by=autos["brand"]).mean().round(decimals=2).to_dict()
```

```python
brand_mean = autos["price_usd"].groupby(by=autos["brand"]).mean().round(decimals=2).to_dict()

for brand, mean in brand_mean.items():
    
    if mean == max(brand_mean.values()):
        print(f"{brand.capitalize()} is the most expensive brand on average, at USD {mean}.")
    
    if mean == min(brand_mean.values()):
        print(f"{brand.capitalize()} is the least expensive brand on average, at USD {mean}.")
```

## Does the average mileage of a brand affect its price?
#### 'brand', 'price_usd', and 'odometer_km'

```python
autos["odometer_km"].groupby(by=autos["brand"]).mean().round(decimals=2).to_dict()
```

```python
brand_mileage = autos["odometer_km"].groupby(by=autos["brand"]).mean().round(decimals=2).to_dict()

for brand, mileage in brand_mileage.items():
    if mileage == max(brand_mileage.values()):
        print(f"{brand.capitalize()} is the most travelled at an average mileage of {mileage} kms")

    if mileage == min(brand_mileage.values()):
        print(f"{brand.capitalize()} is the least travelled at an average mileage of {mileage} kms")
```

```python
avg_price = pd.DataFrame.from_dict(brand_mean, orient='index')
avg_mileage = pd.DataFrame.from_dict(brand_mileage, orient='index')
brand_price_mileage = pd.merge(left=avg_price, right=avg_mileage, how='inner', left_index=True, right_index=True)

brand_price_mileage
```

```python
brand_price_mileage.rename(columns={"0_x": "mean_price", "0_y": "mean_mileage"}, inplace=True)
brand_price_mileage
```
